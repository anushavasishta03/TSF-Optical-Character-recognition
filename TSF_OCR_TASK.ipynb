Importing image processing libraries



```python
from PIL import Image #PIL is python imaging library
import pytesseract
import os
from google.colab.patches import cv2_imshow 
import cv2 #opencv library is imported
```

the image is opened and converted to string


```python
def CR(img_path):
  img = Image.open(img_path)
  image = cv2.imread(img_path)
  cv2_imshow(image)
  converted_string = pytesseract.image_to_string(img, lang='eng')
  print(converted_string)
```

test case 1

Draws rectangle box on each recognized letter (character)

```python
img = cv2.imread('case_1.jpeg')

h,w,c = img.shape
boxes = pytesseract.image_to_boxes(img) 
for b in boxes.splitlines():
    b = b.split(' ')
    img = cv2.rectangle(img, (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])), (0, 255, 0), 2)

cv2_imshow(img)
cv2.waitKey(0)
```
